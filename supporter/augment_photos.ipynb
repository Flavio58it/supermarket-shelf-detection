```python
def random_depth(image_array: ndarray):
    # pick a random degree of rotation between 25% on the left and 25% on the right
    rows,cols,ch = image_array.shape
    
    pts1 = np.float32([[0,0],[rows,0],[0,cols],[rows,cols]])
    pts2_1 = np.float32([[0,0],[rows, cols//8],[0,cols],[rows,cols - cols//8]])
    pts2_2 = np.float32([[0,cols//8],[rows, 0],[0,cols - cols//8],[rows,cols]])
    choose_L_R = random.randint(1, 2)
    if choose_L_R == 1:
        print("PT 1")
        pts2 = pts2_1
    else:
        print("PT 2")
        pts2 = pts2_2

    M = cv2.getPerspectiveTransform(pts1,pts2)

    dst = cv2.warpPerspective(img,M,(cols,rows))
    return dst

def random_noise(image_array: ndarray):
    # add random noise to the image
    print("Random Noise")
    return sk.util.random_noise(image_array)

def horizontal_flip(image_array: ndarray):
    # horizontal flip doesn't need skimage, it's easy as flipping the image array of pixels !
    print("Horizontal Flip")
    return image_array[:, ::-1]

def apply_random_transfomration(available_transformations, 
                                num_transformations_to_apply, 
                                image_to_transform):
    num_transformations = 0
    transformed_image = None
    transformations = []
    while num_transformations < num_transformations_to_apply:
        # choose a random transformation to apply for a single image
        key = random.choice(list(available_transformations))
        transformations.append(key)
        print("Random choice: " + str(key))
        transformed_image = available_transformations[key](image_to_transform)
        num_transformations += 1
        
    return((transformed_image, transformations))
```
